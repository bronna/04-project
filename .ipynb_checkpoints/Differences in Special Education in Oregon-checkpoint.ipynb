```python
import pandas as pd
import altair as alt

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
```

```python
df2 = pd.read_csv("SpEdMediaDistrict-2018-19.csv")
df2.head(1)
```

```python
columns = ['Institution Name', 'Total Student Count', 'LRE Students >80%', 'LRE Students <40%', 'SuspExplFg', 'SuspExplRaceEthnicityFg', 'DisPrptnRprsntnFg', 'DisPrptnRprsntnDsbltyFg', 'IEP 4Yr Cohort Grad 17-18', 'IEP Dropout 17-18']
df = pd.read_csv("SpEdMediaDistrict-2018-19.csv",
                 usecols=columns,
                 na_values=['*'])
df.head()
```

```python
df.info(verbose=True)
```

```python
df['LRE Students >80%'] = df['LRE Students >80%'].str.replace('%','').astype(float)
df['LRE Students <40%'] = df['LRE Students <40%'].str.replace('%','').astype(float)
df['IEP Dropout 17-18'] = df['IEP Dropout 17-18'].str.replace('%','').astype(float)
```

```python
df['LRE Students >80%'].describe()
```

```python
df_large = df[df['Total Student Count'] > 30]
df_large.head()
```

```python
df_large["LRE Students >80%"].describe()
```

```python
df_large["LRE Students <40%"].describe()
```

```python
chart = alt.Chart(df_large).mark_bar().encode(
    alt.X("LRE Students >80%", bin=alt.Bin(maxbins=30)),
    y='count()',
)

line = alt.Chart(pd.DataFrame({'x': [73]})).mark_rule(strokeDash=[10, 10], color="darkorange").encode(x='x')

chart + line
```

```python
number_not_80 = df[df["LRE Students >80%"] < 73].count()["Institution Name"]

number_districts = 196

percent_not_80 = number_not_80 / number_districts

number_not_80, (percent_not_80 * 100).round(2)
```

```python
number_not_40 = df[df["LRE Students <40%"] > 10.6].count()["Institution Name"]

number_districts = 196

percent_not_40 = number_not_40 / number_districts

number_not_40, (percent_not_40 * 100).round(2)
```

```python
alt.Chart(df).mark_point().encode(
#     x='LRE Students >80%',
#     y='IEP 4Yr Cohort Grad 17-18',
    alt.X('LRE Students >80%',
        scale=alt.Scale(zero=False)
    ),
    alt.Y('IEP 4Yr Cohort Grad 17-18',
        scale=alt.Scale(zero=False)
    ),
    size='Total Student Count',
    tooltip=['Institution Name', 'Total Student Count', 'LRE Students >80%', 'IEP 4Yr Cohort Grad 17-18']
).interactive()
```
